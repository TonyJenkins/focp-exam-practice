```python
def contains_coconut(list_of_strings):
    return 'coconut' in list_of_strings

print(contains_coconut(['coconut', 'coconut']))
print(contains_coconut(['dates', 'coconat']))
print(contains_coconut([]))
```

```python
def contains_coconut(list_of_strings):
    for s in list_of_strings:
        if s == 'coconut':
            return True
    return False

print(contains_coconut(['coconut', 'coconut']))
print(contains_coconut(['dates', 'coconat']))
print(contains_coconut([]))
```

```python
list_of_strings = []
while True:
    new_string = input('Enter another string: ')
    if new_string == '':
        break
    else:
        if new_string not in list_of_strings:
            list_of_strings.append(new_string)

print(f'There were {len(list_of_strings)} unique strings.')
```

```python
list_of_strings = set()
while True:
    new_string = input('Enter another string: ')
    if not new_string:
        break
    else:
        list_of_strings.add(new_string)

print(f'There were {len(list_of_strings)} unique strings.')
```
